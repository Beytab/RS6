```python
import pandas as pd 
from sklearn.feature_extraction import DictVectorizer
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
train_data = pd.read_csv('C:/Users/18280/Desktop/RS/主课/assignment/train.csv')
test_data = pd.read_csv('C:/Users/18280/Desktop/RS/主课/assignment/test.csv')
```

```python
train_data.head()
```

```python
train_data.info()
```

```python
#数据集拆分
train_x,validation_x,train_y,validation_y = train_test_split(train_data,
                                                             train_data.Attrition,
                                                             test_size=0.25,
                                                             random_state=33)
# 特征选择
train_features = train_x[train_x.columns.difference(['user_id','Attrition'])]
train_labels = train_x['Attrition']
validation_features = validation_x[validation_x.columns.difference(['user_id','Attrition'])]
validation_labels = validation_x['Attrition']
test_features = test_data[test_data.columns.difference(['user_id','Attrition'])]
# print(validation_features.head())
# print('_'*30)
# print(train_features.head())
# print('-'*30)
# print(train_labels.head())
# print('-'*30)t
print(validation_labels.head())
# print(train_features.to_dict(orient='records')[1])
# print(validation_labels.to_dict(orient='records'))
```

```python
# 将非数字化数据数值化
dvec = DictVectorizer(sparse=True)
train_features = dvec.fit_transform(train_features.to_dict(orient='records'))
# train_labels =dvec.transform(train_labels.to_dict(orient='records'))
# train_labels =dvec.transform(train_labels.to_dict())
validation_features = dvec.transform(validation_features.to_dict(orient='records'))
# validation_labels = dvec.transform(validation_labels.to_dict(orient='records'))
# validation_labels = dvec.transform(validation_labels.to_dict())
test_features = dvec.transform(test_features.to_dict(orient='records'))
```

```python
# 采用Z-score规范化
ss = preprocessing.StandardScaler(with_mean=False)
train_features = ss.fit_transform(train_features)
validation_features = ss.transform(validation_features)
test_features = ss.transform(test_features)
```

```python
lr = LogisticRegression()
lr.fit(train_features,train_labels)
```

```python
pre_labels = lr.predict(validation_features)
acc_lr = round(accuracy_score(validation_labels,pre_labels),6)
print('准确率：%.4lf'%acc_lr)
```
