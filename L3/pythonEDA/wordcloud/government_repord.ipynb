```python
import jieba
import wordcloud
import matplotlib.pyplot as plt

f = open('2019.txt','r',encoding='utf-8')
s = f.read()
f.close()
ls = jieba.lcut(s)
txt = ' '.join(ls)
stop_words = {'','','','','','','','','',''}
w = wordcloud.WordCloud(font_path='msyh.ttc',stopwords=stop_words,\
                       width=1000,height=800,background_color='white')
w.generate(txt)
# w.to_file('goverment.jpg')
plt.imshow(w)
plt.show()
```
