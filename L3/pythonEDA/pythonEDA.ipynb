```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from matplotlib.font_manager import FontProperties
```

```python
# 散点图
def scatter():
    # 数据准备
    N = 500
    x = np.random.randn(N)
    y = np.random.randn(N)
    # 用Matplotlib画散点图
    plt.scatter(x,y,marker='x')
    plt.show()
    # 用Seaborn画散点图
    df = pd.DataFrame({'x':x,'y':y})
    sns.jointplot(x='x',y='y',data=df,kind='scatter')
    plt.show()
    
scatter()
```

```python
# 折线图
def line_chart():
    x = [1900, 1901, 1902, 1903, 1904, 1905, 1906, 1907, 1908, 1909, 1910]
    y = [265, 323, 136, 220, 305, 350, 419, 450, 560, 720, 830]
    # 使用Matplotlib画折线图
    plt.plot(x, y)
    plt.show()
    # 使用Seaborn画折线图
    df = pd.DataFrame({'x':x,'y':y})
    sns.lineplot(x='x',y='y',data=df)
    plt.show()
    
line_chart()
```

```python
#条形图
def bar():
    x = ['c1', 'c2', 'c3', 'c4']
    y = [15, 18, 5, 26]
    plt.bar(x,y)
    plt.show()
    
    #seaborn实现
    sns.barplot(x,y)
    plt.show()
bar()
```

```python
#饼图
def pie_char():
    num = [25,33,37]
    # 射手adc：法师apc：坦克tk
    labels = ['ADC','APC', 'TK']
    # 用Matplotlib画饼图
    plt.pie(x=num,labels=labels)
    plt.show()
def pie_char2():
    data = {}
    data['ADC'] = 25
    data['APC'] = 33
    data['TK'] = 37
    data = pd.Series(data)
    data.plot(kind='pie',labels='heros')
    plt.show()
pie_char()
pie_char2()
```

```python
# 热力图
def thermodynamic():
    np.random.seed(33)
    data = np.random.rand(3,3)
    heatmap = sns.heatmap(data)
    plt.show()
def thermodynamic2():
    flights = sns.load_dataset('flights')
    print(flights.head(10))
    flights = flights.pivot('month','year','passengers')
    print('-'*100)
    print(flights.head())
    sns.heatmap(flights)
    sns.heatmap(flights,linewidth=.5,annot=True,fmt='d')
    plt.show()
    
thermodynamic()
thermodynamic2()
```

```python
# 箱线图
def box_plot():
    # 生成0-1之间的10*4维度数据
    data = np.random.normal(size=(10,4))
    labels = ['A','B','C','D']
    plt.boxplot(data,labels=labels)
    plt.show()
    
    #seaborn
    df = pd.DataFrame(data,columns=labels)
    sns.boxplot(data=df)
    plt.show()
    
box_plot()
```
