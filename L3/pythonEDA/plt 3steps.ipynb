```python
import random
import matplotlib.pyplot as plt
# import matplotlib
# matplotlib.rcParams['font.family'] = 'SimHei'
x = range(60)
y_shanghai = [random.uniform(16,18) for i in x]

plt.figure(figsize=(15,6),dpi=130)
plt.plot(x,y_shanghai)

#修改x,y axis
#准备x_label
x_label = ['11点{}分'.format(i) for i in x]
plt.xticks(x[::5],x_label[::5],fontproperties='SimHei')
plt.yticks(range(0,30,5))
#增加网格,alpha透明度
plt.grid(True,linestyle='--',alpha=0.8)
#添加描述信息
plt.xlabel('时间',fontproperties='SimHei',fontsize=20)
plt.ylabel('温度',fontproperties='SimHei',fontsize=20)
plt.title('时间温度变化曲线',fontproperties='SimHei',fontsize=20)

plt.show()
```

```python
import random
import matplotlib.pyplot as plt
# import matplotlib
# matplotlib.rcParams['font.family'] = 'SimHei'
x = range(60)
y_shanghai = [random.uniform(16,18) for i in x]
y_beijing = [random.uniform(1,10) for i in x]

plt.figure(figsize=(15,6),dpi=130)
plt.plot(x,y_shanghai,color='b',linestyle='-.',marker='o',label='SH')
plt.plot(x,y_beijing,color='r',linestyle='--',marker='*',label='BJ')

#显示图例
plt.legend()

#修改x,y axis
#准备x_label
x_label = ['11点{}分'.format(i) for i in x]
plt.xticks(x[::5],x_label[::5],fontproperties='SimHei')
plt.yticks(range(0,30,5))
#增加网格,alpha透明度
plt.grid(True,linestyle='--',alpha=0.8)
#添加描述信息
plt.xlabel('时间',fontproperties='SimHei',fontsize=20)
plt.ylabel('温度',fontproperties='SimHei',fontsize=20)
plt.title('时间温度变化曲线',fontproperties='SimHei',fontsize=20)

plt.show()
```

# 多个坐标系显示
### 采用面向对象的方法

```python
import random
import matplotlib.pyplot as plt
# import matplotlib
# matplotlib.rcParams['font.family'] = 'SimHei'
# 1.准备数据
x = range(60)
y_shanghai = [random.uniform(16,18) for i in x]
y_beijing = [random.uniform(1,10) for i in x]

# 2.创建画布
# plt.figure(figsize=(15,6),dpi=130)
figure,axis = plt.subplots(nrows=1,ncols=2,figsize=(15,6),dpi=150)

# 3.绘制图像
axis[0].plot(x,y_shanghai,color='b',linestyle='-.',marker='o',label='SH')
axis[1].plot(x,y_beijing,color='r',linestyle='--',marker='*',label='BJ')

#显示图例
axis[0].legend()
axis[1].legend()


#修改x,y axis
#准备x_label
x_label = ['11点{}分'.format(i) for i in x]
# axis[0].set_xticks(x[::5],x_label[::5],fontproperties='SimHei')
axis[0].set_xticks(x[::10])
axis[0].set_xticklabels(x_label[::10],fontproperties='SimHei')
axis[0].set_yticks(range(0,30,5))
# axis[1].set_xticks(x[::5],x_label[::5],fontproperties='SimHei')
axis[1].set_xticks(x[::10])
axis[1].set_xticklabels(x_label[::10],fontproperties='SimHei')
axis[1].set_yticks(range(0,30,5))
#增加网格,alpha透明度
axis[0].grid(True,linestyle='--',alpha=0.8)
axis[1].grid(True,linestyle='--',alpha=0.8)
#添加描述信息
axis[0].set_xlabel('时间',fontproperties='SimHei',fontsize=20)
axis[0].set_ylabel('温度',fontproperties='SimHei',fontsize=20)
axis[0].set_title('上海时间温度变化曲线',fontproperties='SimHei',fontsize=20)
axis[0].set_xlabel('时间',fontproperties='SimHei',fontsize=20)
axis[0].set_ylabel('温度',fontproperties='SimHei',fontsize=20)
axis[0].set_title('北京时间温度变化曲线',fontproperties='SimHei',fontsize=20)

plt.show()
```
