```python
from wordcloud import WordCloud
import pandas as pd
import matplotlib.pyplot as plt

def data_load():
    data = pd.read_csv('Market_Basket_Optimisation.csv',header=None)
    data = data.fillna(0)
    transactions = []
    for index,row in data.iterrows():
        for item in row:
            if item!=0:
                transactions.append(item)
    return transactions

def create_word_cloud(f):
    print('开始绘制词云')
    txt = " ".join(f)
    w = WordCloud(width=1000,height=700,\
                  background_color='white',max_words=10
                 )
    w.generate(txt)
    w.to_file('Market_wordcloud.png')
    plt.imshow(w)
    plt.show()

transactions = data_load()
create_word_cloud(transactions)
```
