整体架构如图所示：
* 推荐系统可以分为召回和排序两个部分
* 召回部分根据用户画像和实时的场景数据返回相关度较高的资源作为候选集
* 排序阶段通过更加复杂精细的特征对候选集进行排序,最终呈现给用户推荐结果

## 在召回阶段中
* 输入的是用户的最近30天的听歌历史、搜索历史和人口统计学信息以及近30分钟的信息作为输入向量(去掉听歌时间小于30s的样本)
* DNN的隐藏层可以选择三层1024ReLu → 512ReLu → 256Relu塔型结构
* 输出层采用分为线上和离线部分:线上部分可以采用ANN的方式进行分类,离线部分可以用softmax计算最终分类

## 排序阶段中
### 输入部分
* 可以用更加精细化的特征:歌曲部分的化可以采用NLP的方式提取特征(主要元素为词,曲,配乐),歌曲的热度等
* 还可以有用户与音乐的关系、用户画像、用户本身的固有喜好(偏好民谣还摇滚还流行),
* 上下文的场景

### 隐藏层结构依然可以采用三层1024ReLu → 512ReLu → 256Relu塔型结构

### 输出阶段:可以根据单首歌曲听歌次数进行加权,正样本权重为听歌的次数,负样本不加权,所以可以采用可以Weighed LR放模拟输出
