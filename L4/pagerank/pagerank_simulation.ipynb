```python
import numpy as np

#设置转移矩阵
m = np.array(
             [[0,0,0,1/3,0,0],
             [1/4,0,0,0,1/2,0],
             [0,1,0,1/3,1/2,0],
             [1/4,0,0,0,0,1],
             [1/4,0,1,1/3,0,0],
             [1/4,0,0,0,0,0]])
#初始化向量
w = np.array([1/6,1/6,1/6,1/6,1/6,1/6])

#简单PageRank
def work(w,m,n):
    for i in range(n):
        w = np.dot(m,w)
        print(w)
        
#pagerank随机浏览模型
def random_work(w,m,n):
    d = 0.85 #阻尼因子
    for i in range(n):
        w = 1-d/6+d*np.dot(m,w)
        print(w)

work(w,m,100)
print('-'*100)
random_work(w,m,100)
```
