```python
import jieba
import jieba.analyse
import jieba.posseg as pseg

#读取文本
f = open('news.txt','r',encoding='utf-8')
s = f.read()
f.close()
#获取分词
seg_list = jieba.cut(s,cut_all=False)
# 获取分词和词性
words = pseg.cut(s)
print('词性：')
for word,tag in words:
    print('%s,%s'%(word,tag))
print('-'*100)

# 通过TF-IDF获取关键词
keywords_tf = jieba.analyse.extract_tags(s,topK=20,withWeight=True,allowPOS=('n','nr','ns','v'))
print('IF-IDF关键词提取结果：')
for item in keywords_tf:
    print(item[0],item[1])
print('_'*100)

#基于textrank关键词提取
keywords_t = jieba.analyse.textrank(s,topK=20,withWeight=True,allowPOS=('n','nr','ns','v'))
for item in keywords_t:
    print(item[0],item[1])
```

```python
# 摘要提取
from textrank4zh import TextRank4Sentence

tr4s = TextRank4Sentence()
tr4s.analyze(text=s,lower=True,source='all_filters')

abstract = {}
for item in tr4s.get_key_sentences(num=5):
    abstract[item.index] = item.sentence
    
print(abstract.values())
```
